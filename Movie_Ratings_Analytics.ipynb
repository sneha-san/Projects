# 13) scatter plot of critics rating vs audience rating 

```python
plt.scatter(x='Rotten Tomatoes Ratings %',y='Audience Ratings %',data=ad_data)
plt.xlabel('critic rating')
plt.ylabel('audience rating')
```

# 14) box plot based on genre and critics rating 

```python
sns.boxplot(x="Genre",y="Rotten Tomatoes Ratings %",data=ad_data,palette='rainbow')
```

# 15) box plot based on genre and audience rating 

```python
sns.boxplot(x="Genre",y="Audience Ratings %",data=ad_data,palette='rainbow')
```

# 16) pairplot of the whole data set 

```python
sns.pairplot(ad_data)
```

# 19) Joint Plot of critics rating and audience rating (with type as hexagonal) 

```python
sns.jointplot(x='Rotten Tomatoes Ratings %',y='Audience Ratings %',data=ad_data,kind='hex')
```

# 20) Joint plot of budget vs genre (type as scatter) 


```python
sns.jointplot(x='Budget (million $)',y='Genre',data=ad_data,kind='scatter')
```
